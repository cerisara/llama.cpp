```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def swiglu(x):
    return (x**2)*sigmoid(1.14776*x)

def rescaled(x, strength):
    return (x**2)*sigmoid(strength*x)

def rescaled_biased(x, strength, treshold):
    return (x+(1 / ((1-treshold) * sigmoid(strength*(1-treshold)))) - 1)*(x-treshold)*sigmoid(strength*(x-treshold))

def weird(x, strength):
    return (x+(1 / (sigmoid(strength)) - 1))*x*sigmoid(strength*x)

def weird2(x, strength, bias1):
    return (x+bias1)*(x-0.2)*sigmoid(strength*x -0.2)
```

```python
x = np.linspace(-1, 1, 1000)

plt.plot(x, swiglu(x), label='swiglu')
# plt.plot(x, rescaled(x, 50), label='rescaled 50')
# plt.plot(x, rescaled_biased(x, 50, 0.5), label='rescaled_biased 50 0.5')
# plt.plot(x, rescaled(x, i), label='rescaled '+str(i))
plt.plot(x, weird(x, 0), label='weird 0 1')
plt.plot(x, weird(x, 1), label='weird 1 1')
# plt.plot(x, weird(x, 50), label='weird 50 1')
plt.plot(x, weird2(x, 1, 1), label='weird2 1 1')
# plt.plot(x, weird2(x, 50, 1), label='weird2 50 1')
# plt.plot(x, bias1(x), label='bias1')
plt.vlines(0, -1, 1, linestyles='dashed', linewidth=0.2)
plt.hlines(0, -1, 1, linestyles='dashed', linewidth=0.2)
plt.legend()
plt.show()
```
